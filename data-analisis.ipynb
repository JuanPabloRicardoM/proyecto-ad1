La categoria seria la variable objetivo, saber si es Fraude-y-Practicas-Indebidas, se ve que tiene muchos tipos de datos esta variables pero es porque tiene categorias en mayuscula y minuscula para referirse al mismo texto

```python
df_historico['Categoria'] = df_historico['Categoria'].str.replace('Sin_Hallazgo', 'SIN_HALLAZGO')
df_historico['Categoria'] = df_historico['Categoria'].str.replace('Otras_Políticas', 'OTRAS_POLITICAS')
df_historico['Categoria'] = df_historico['Categoria'].str.replace('Limitación_en_el_Alcance', 'LIMITACIÓN_EN_EL_ALCANCE')
df_historico['Categoria'] = df_historico['Categoria'].str.replace('Controles_Administrativos_de_Agencia', 'CONTROLES_ADMINISTRATIVOS_DE_AGENCIA')
df_historico['Categoria'] = df_historico['Categoria'].str.replace('Políticas_Críticas', 'POLITICAS_CRITICAS')
df_historico['Categoria'] = df_historico['Categoria'].str.replace('Estabilidad', 'ESTABILIDAD')

df_historico['Categoria'] = df_historico['Categoria'].str.replace('Calidad_de_Datos', 'CALIDAD_DE_DATOS')
df_historico['Categoria'] = df_historico['Categoria'].str.replace('otros', 'Otros')
df_historico['Categoria'] = df_historico['Categoria'].str.replace('Reputación', 'REPUTACIÓN')
df_historico['Categoria'] = df_historico['Categoria'].str.replace('Habilidad_Empresarial', 'HABILIDAD_EMPRESARIAL')
df_historico['Categoria'] = df_historico['Categoria'].str.replace('Existencia_y_Continuidad_del_Negocio', 'EXISTENCIA_Y_CONTINUIDAD_DEL_NEGOCIO')
df_historico['Categoria'] = df_historico['Categoria'].str.replace('Fraude_y_Prácticas_Indebidas', 'FRAUDE_Y_PRÁCTICAS_INDEBIDAS')

df_historico['Categoria'] = df_historico['Categoria'].str.replace('Flujo_de_Caja', 'FLUJO_DE_CAJA')
df_historico['Categoria'] = df_historico['Categoria'].str.replace('Incumplimiento_Al_SAC', 'INCUMPLIMIENTO_AL_SAC')
```

```python
df_historico['Categoria'].value_counts()
```

Podemos borrar columnas categoricas unicas, como por ejemplo Num, ID_modif, Cliente, Nombre Analista, Validacion unico, Borramos nombre de analista porque tenemos la columna Analista

```python
df_clean = df_historico.drop(['Cliente', 'Num', 'ID_modif', 'Nombre analista', 'Validación unico'], axis=1)
df_clean['Cuota'] = df_clean[['Cuota']].replace(' $ -   ', np.nan)
df_clean['Cuota'].astype('float')
df_clean.info()
```
